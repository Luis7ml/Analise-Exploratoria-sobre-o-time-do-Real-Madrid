```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('Tabela_Real_Madrid.csv', sep=',')
dados.columns = dados.columns.str.strip()
dados
```

```python
dados.describe()
```

```python
dados = dados[['Season', 'LgRank', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'Pts', 'Top Team Scorer']]
campeoes = dados.loc[dados['LgRank'] == '1st']
campeoes
artilheros = campeoes['Top Team Scorer']
campeoes.plot(x='Top Team Scorer', y='GF', kind='bar', color='red')
```

```python
dados.plot(x='Season', y='W', kind='barh', color='Blue')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
